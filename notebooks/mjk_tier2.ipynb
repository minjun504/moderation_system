```python
import torch
from torch import nn
from transformers import DistilBertTokenizer
```

```python
class DistilBertRegressor(nn.Module):
    def __init__(self, vector_size):
        super().__init__()
        self.distilbert = DistilBertTokenizer.from_pretrained("distilbert-base-uncased")
        self.regressor = nn.Linear(in_features=vector_size, out_features=1)
    
    def forward(self, id, attention):
        output = self.distilbert(input_ids=id, attention_mask=attention)
        cls_vec = output.last_hidden_state[:, 0, :]
        score = self.regressor(cls_vec)
        return score
```
